### Dummy Variable Creation

```python
allDummies = pd.get_dummies(mercedesBenz[['X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X8']])
# Drop one column from each to prevent multicollinearity
allDummies.drop(['X0_a', 'X1_a', 'X2_a', 'X3_a', 'X4_a', 'X5_aa', 'X6_a', 'X8_a'], axis=1, inplace=True)
# Now concatenate to the original dataframe and drop the original columns, and ID because it's useless
# Also save the target variable and then drop it
mercedesBenz = pd.concat([mercedesBenz, allDummies], axis=1)
testingTime = mercedesBenz['y']
mercedesBenz.drop(['ID', 'y', 'X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X8'], axis=1, inplace=True)
mercedesBenz.head()
```

```python
testMercBenz = pd.read_csv('./test.csv')
allDummies = pd.get_dummies(testMercBenz[['X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X8']])
# Drop one column from each to prevent multicollinearity
allDummies.drop(['X0_a', 'X1_a', 'X2_a', 'X3_a', 'X4_a', 'X5_aa', 'X6_a', 'X8_a'], axis=1, inplace=True)
# Now concatenate to the original dataframe and drop the original columns, and ID because it's useless
# Also save the target variable and then drop it
testMercBenz = pd.concat([testMercBenz, allDummies], axis=1)
testID = testMercBenz['ID']
testMercBenz.drop(['ID', 'X0', 'X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X8'], axis=1, inplace=True)
testMercBenz.head()
```

```python
# There are columns in the test data that are not in the training data. Add these
columnsToAdd = list(filter(lambda x: x not in mercedesBenz.columns, testMercBenz.columns))
columnsToAdd = DataFrame(data=np.zeros((len(mercedesBenz), len(columnsToAdd)), dtype=int), columns=columnsToAdd)
mercedesBenz = mercedesBenz.join(columnsToAdd)
mercedesBenz.head()
```

```python
# The same is true for the test data
columnsToAdd = list(filter(lambda x: x not in testMercBenz.columns, mercedesBenz.columns))
columnsToAdd = DataFrame(data=np.zeros((len(testMercBenz), len(columnsToAdd)), dtype=int), columns=columnsToAdd)
testMercBenz = testMercBenz.join(columnsToAdd)
testMercBenz.head()
```

## Dimensionality Reduction
There are multiple ways to reduce the number of dimensions. Most common is Principal Component Analysis (PCA). I will try a few and see how they compare when predicting the testing time of the car.
### Principal Component Analysis (PCA)
PCA needs the values to all be standardized because it is sensitive to centering.

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
standardizedMercedesBenz = preprocessing.scale(mercedesBenz)
standardizedTestingTime = preprocessing.scale(testingTime)
```

```python
pcaMod = PCA()
pcaMod.fit(mercedesBenz)
```

```python
# View the explained ratio.

temp = DataFrame(columns=['axis', 'Explained Variance Ratio'])
temp['axis'] = np.arange(1, len(pcaMod.explained_variance_ratio_) + 1)
temp['Explained Variance Ratio'] = pcaMod.explained_variance_ratio_

fig = plt.figure(figsize=(12, 6), dpi= 100)
sns.lmplot(x='axis', y='Explained Variance Ratio', data=temp, fit_reg=False)
```

Okay so it appears that we don't need to keep anything with a variance ratio of lower than 0.01% as it starts to slowly flat line from there.

```python
amountToKeep = len(pcaMod.explained_variance_ratio_[pcaMod.explained_variance_ratio_ > 0.0001])
amountToKeep
```

Make the model again only this time put n_components to be 342. I am doing it again for convenience of transforming back and forth.

```python
pcaMod = PCA(n_components=amountToKeep)
reducedX = pcaMod.fit_transform(mercedesBenz)
```

Now get the reduced components and run regression with these components against the testing time.

```python
reducedX = DataFrame(data=reducedX, columns=['P' + str(i) for i in range(1, reducedX.shape[1] + 1)])
```

```python
reducedX.head()
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
pcaLinReg = LinearRegression()
pcaLinReg.fit(reducedX, testingTime)
```

The coefficients are with respect each of the reduced number of transformed variables. We have to transform this back into the originial to get a matrix of original dimensions.

```python
pcaMod.components_.shape
```

```python
pcaLinReg.coef_.shape
```

```python
origXCoef = np.dot(pcaLinReg.coef_, pcaMod.components_) + pcaMod.mean_
```

```python
mercedesBenz.head()
```

```python
predictions = np.dot(mercedesBenz, origXCoef)
predictions
```

```python
testingTime[:5]
```

```python
from scipy.stats import f
ssm = np.sum((predictions - np.mean(standardizedTestingTime)) ** 2)
sse = np.sum((standardizedTestingTime - predictions) ** 2)
dfm = len(standardizedMercedesBenz[0]) - 1
dfe = len(standardizedMercedesBenz) - len(standardizedMercedesBenz[0])
MSM = ssm / dfm
MSE = sse / dfe
```

```python
F = MSM / MSE
1 - f.cdf(F, dfm, dfe)
```

```python
MSE
```

```python
MSM
```

### Submitting the Test File

```python
# Do PCA using the parameters we made before.
pcaModTestX = PCA(n_components=amountToKeep)
reducedTestX = pcaModTestX.fit_transform(testMercBenz)
# Get original coefficients
origTestXCoef = np.dot(pcaLinReg.coef_, pcaModTestX.components_) + pcaModTestX.mean_
# Make predictions
predictionsTestX = np.dot(testMercBenz, origTestXCoef)
predictionsTestX[:5]
```

```python
# Now save these as DataFrame, write to a csv and submit!
results = DataFrame(columns=['ID', 'y'])
results['ID'] = testID
results['y'] = predictionsTestX
```

```python
results.to_csv('./resultsPCA.csv', index=False)
```

## Neural Networks
I'm going to try Python's built-in neural network method then I'm going to try something like Tensorflow.
### Sci-kit learn's Multi-Layer Perceptron

```python
mercedesBenz.head()
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
# As always, scale the features. For meaninful results,
# the same transformation must be applied to the test.
XScalar = StandardScaler()
XScalar.fit(mercedesBenz)
scaledXTrain = XScalar.transform(mercedesBenz)
```

```python
# Use GridSearchCV to find the best regularization parameter.
from sklearn.model_selection import GridSearchCV
neuralReg = MLPRegressor(solver='adam', hidden_layer_sizes=(10, 10, 10), random_state=1)
parameters = {'alpha': 10.0 ** -np.arange(1, 7)}
neuralReg = GridSearchCV(neuralReg, parameters)
neuralReg.fit(scaledXTrain, testingTime)
```

```python
# Do an F test for meaningful regression
predictions = neuralReg.predict(scaledXTrain)
SSM = np.sum((predictions - np.mean(testingTime)) ** 2)
SSE = np.sum((testingTime - predictions) ** 2)
DFM = len(mercedesBenz.columns) - 1
DFE = len(mercedesBenz) - len(mercedesBenz.columns)
MSM = SSM / DFM
MSE = SSE / DFE
F = MSM / MSE
1 - f.cdf(F, DFM, DFE)
```

Looks like the regression is significant, now let's submit.

```python
# Apply same scaling
scaledXTest = XScalar.transform(testMercBenz)
# Use it to predict the testing time
predictions = neuralReg.predict(scaledXTest)
results = DataFrame(columns=['ID', 'y'])
results.ID = testID
results.y = predictions
```
